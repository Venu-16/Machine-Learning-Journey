```python
data = pd.read_csv('/content/2019-05-31-data (1).csv')
view_cnt = data['view_count']
like_cnt = data['likes']
ratio = data['ratio']

plt.scatter(view_cnt,like_cnt, c = ratio, cmap='summer',
            edgecolors='black',linewidths=1, alpha=0.75)
cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')


plt.tight_layout()
plt.show()
```
