**Filling Area on line Plots**

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/data (1).csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']


plt.plot(ages,dev_salaries,color='#444444',
            linestyle='--',label='All Devs')

overall_median = 57287

plt.plot(ages, py_salaries,label='Python')

# plt.fill_between(ages,py_salaries, where=(py_salaries>overall_median),interpolate=True,alpha=0.25)
# plt.fill_between(ages,py_salaries, where=(py_salaries<=overall_median),color = 'r',interpolate=True,alpha=0.25)

plt.fill_between(ages,py_salaries,dev_salaries, where=(py_salaries>overall_median),interpolate=True,alpha=0.25 , label = 'Above avg')
plt.fill_between(ages,py_salaries,dev_salaries, where=(py_salaries<=overall_median),color = 'r',interpolate=True,alpha=0.25, label = 'Below avg')



plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.show()
```
